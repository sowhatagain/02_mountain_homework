```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('Mountains.csv',  delimiter = ',')
#data.head(10)
#data.info()
```

```python
data.loc[pd.isnull( data['Parent mountain'] ) ].head() # нет названия родителя (1 entry is null)
data['Parent mountain'].fillna("Main", inplace = True ) # для удобства работы заменим на Main

only_mountain = {'Mountain':data['Mountain'], 'Parent':data['Parent mountain']} #работаем только с названиями гор
data_mountains= pd.DataFrame(only_mountain)
#data_mountains.head(10)
#data_mountains.info()

data_mountains['Amount'] = 1            #флаг для count() group by 
data_mountains.head(3)
```

```python
parent_mountains = data_mountains.groupby('Parent').sum().reset_index() # группируем по количеству родителей
#parent_mountains.head()

mask = (parent_mountains.Amount > 3)          # отбираем с числом вершин больше трех
mountains_over_three = parent_mountains.loc[ mask ].copy()

#выведем в файл
mountains_over_three.to_csv( 'Mountains_groups_over_three_count.csv', sep = ',', na_rep = '0', columns = ['Parent','Amount'], index = True )

mountains_over_three.sort_values( by = 'Amount', ascending = False ).head(10)
```

```python
print(mountains_over_three['Amount'].sum()) # проверка сколько записей о горах будет
```

```python
intersection = data.merge( mountains_over_three, how = 'inner', left_on = 'Parent mountain', right_on = 'Parent' )
#intersection.info() # все ок
```
